```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import date
import matplotlib.pyplot as plt
```

```python
ensign = pd.read_csv('ensign_snf/ensign_ca_2020_to_2021.csv')
all_ca = pd.read_csv('all_us_snf/all_ca_2020_to_2021.csv')
```

```python
all_ca_2020 = all_ca[all_ca['year'] == 2020]
all_ca_2021 = all_ca[all_ca['year'] == 2021]
```

```python
ensign_2020 = ensign[ensign['year'] == 2020]
ensign_2021 = ensign[ensign['year'] == 2021]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
all_ca_2020.hist('occupancy', bins=np.arange(0, 1 + 0.1, 0.05) , ax=axes[0])
axes[0].set_title('2020 CA occupancy')
axes[0].set_xlabel('certified bed occupancy')
all_ca_2021.hist('occupancy', bins=np.arange(0, 1 + 0.1, 0.05) , ax=axes[1])
axes[1].set_title('2021 CA occupancy')
axes[1].set_xlabel('certified bed occupancy')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
ensign_2020.hist('occupancy', bins=np.arange(0, ensign_2020['occupancy'].max() + 0.1, 0.05) , ax=axes[0])
axes[0].set_title('2020 Ensign CA occupancy')
axes[0].set_xlabel('certified bed occupancy')
ensign_2021.hist('occupancy', bins=np.arange(0, ensign_2021['occupancy'].max() + 0.1, 0.05) , ax=axes[1])
axes[1].set_title('2021 Ensign CA occupancy')
axes[1].set_xlabel('certified bed occupancy')
```
